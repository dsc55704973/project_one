```python
# Dependencies

import json
import requests
from config import api_key
import matplotlib.pyplot as plt
from scipy import stats
import pandas as pd
import numpy as np
from csv import writer
from csv import reader
```

```python
# sector_list=["Energy", "Consumer Discretionary", "Information Technology", "Industrials", "Communication Services", "Financials", "Consumer Staples", "Health Care", "Real Estate", "Utilities"]
sector_dict={"1-Day": '', "5-Day": '', "1-mo": '', "3-mo": '', "YTD": '', "1-yr": '', "3-yr": ''}
# series=pd.Series(sector_dict, index =['Energy', 'Informatin Technology', 'Consumer Discretionary', 'Health Care'])
df=pd.DataFrame([sector_dict], index=['Energy', 'Information Technology', 'Consumer Discretionary', 'Health Care'])

df
```

```python
# as of 8/4 at 1:02p
url = "https://www.alphavantage.co/query?"
query_url = f"{url}function=SECTOR&apikey={api_key}"
response=requests.get(query_url).json()
response
for index, row in df.iterrows():
    df.loc[index, "1-Day"]=response["Rank B: 1 Day Performance"][index]
    df.loc[index, "5-Day"]=response["Rank C: 5 Day Performance"][index]
    df.loc[index, "1-mo"]=response["Rank D: 1 Month Performance"][index]
    df.loc[index, "3-mo"]=response["Rank E: 3 Month Performance"][index]
    df.loc[index, "YTD"]=response["Rank F: Year-to-Date (YTD) Performance"][index]
    df.loc[index, "1-yr"]=response["Rank G: 1 Year Performance"][index]
    df.loc[index, "3-yr"]=response["Rank H: 3 Year Performance"][index]
df.to_csv("Returns.csv")
df
```
